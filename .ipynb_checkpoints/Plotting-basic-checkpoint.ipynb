### Plotting with Python
There is a variety of different package that can be used for plotting in python. The most common are:
* matplotlib
* seaborn 
* bokeh

In this tutorial we will dive into matplotlib functionality and also into matplotlib directly integrated into pandas.

#### Importing package

```python
import matplotlib.pyplot as plt
```

https://matplotlib.org/3.2.1/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py

Matplotlib has two different approaches to create a plot:
* functional

Rely on pyplot to automatically create and manage the figures and axes, and use pyplot functions for plotting. It is similar to MATLAB plotting style
* object-oriented

Explicitly create figures and axes, and call methods on them (the "object-oriented (OO) style").

https://matplotlib.org/3.2.1/tutorials/introductory/usage.html#figure

```python
import numpy as np
x = np.linspace(0, 20, 20)
```

```python
plt.plot(x, x, label='linear')
```

```python
#MATLAB functional style
import numpy as np
x = np.linspace(0, 20, 20)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Here comes the title")
plt.legend()
```

```python
#nrows, ncols, number of the plot
plt.subplot(1, 3, 1)
plt.plot(x, x, label='linear')

plt.subplot(1, 3, 2)
plt.plot(x, x**2, label='quadratic', color='orange')

plt.subplot(1, 3, 3)
plt.plot(x, x**3, label='cubic', color='green')
```

The OOB style gives you more control over the figure. In the object-oriented interface the plotting functions are methods of explicit Figure and Axes objects. To re-create the previous plot using this style of plotting, you might do the following:

```python
# OOB style
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Here comes the title")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

Adjusting the color is made easy using the **color** parameter inside the plot() method, color can be specified as 
* name color='blue'
* grey scale between 0 and 1 color='0.5'
* short color code color='r'
* Hex code color='#800000'
* RGB tuple, values 0 to 1 color=(0.8,0.2,0.3)

```python
plt.plot(x, x, color='blue')        # specify color by name
plt.plot(x, x**2, color='r')           # short color code (rgbcmyk)
plt.plot(x, x+100, color='0.5')        # Grayscale between 0 and 1
plt.plot(x, x-100, color='#800000')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, x**3, color=(0.8,0.2,0.3)) # RGB tuple, values 0 to 1
```

```python
plt.plot(x, x, color='blue')   
plt.xlim((0,10))
```

```python
# Scatter plot
plt.plot(x, x**3, 'o')
```

### Plotting in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('export-EtherPrice.csv')
```

```python
df.Value.plot()
```

```python
df.Value.cumsum().plot(grid=False)
```

```python
from sklearn.datasets import load_iris
data = load_iris(return_X_y=False)
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = pd.Series(data.target)
```

```python
df.drop('species', axis=1).plot()
```

```python
df.drop('species', axis=1).cumsum(0).plot(subplots=True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))
for i,var in enumerate(df.drop('species', axis=1).columns):
    df.drop('species', axis=1)[var].plot(ax=axes[i], title=var)
axes[0].set_ylabel('observation')
```

### Bar plots 
Bar plots are useful for displaying and comparing measurable quantities, such as counts or volumes. In Pandas, we just use the plot method with a kind='bar' argument.

```python
df.groupby('species').sum().plot(kind='bar')
```

```python
df.groupby('species').sum().plot(kind='bar')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

### Histograms

Histogram is a graph that displays relative frequencies of data values. This can either be raw counts of values or scaled proportions

```python
df.iloc[:,0].hist(grid=False)
```

```python
df.iloc[:,0].hist(bins=30)
```

### Boxplots
Another way of plotting a distribtuion is a boxplot

```python
plt.figure(figsize=(20,10))
df.boxplot( grid=False)
```
